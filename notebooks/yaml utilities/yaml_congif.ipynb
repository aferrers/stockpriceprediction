```python
import os
import yaml

# Define the root directory of your project
root_dir = '../'

# Function to recursively get all subdirectories
def get_subdirectories(directory):
    subdirectories = [os.path.join(directory, d) for d in os.listdir(directory) if os.path.isdir(os.path.join(directory, d))]
    for subdir in subdirectories:
        print(subdir)
        subdirectories.extend(get_subdirectories(subdir))
    return subdirectories

# Get all subdirectories in the project
subdirectories = get_subdirectories(root_dir)

# Convert the list of directories to a dictionary
config = {'folders': {'root_dir': root_dir, 'subdirectories': subdirectories}}

# Write the configuration to a YAML file
with open('../config.yaml', 'w') as file:
    yaml.dump(config, file)
```
